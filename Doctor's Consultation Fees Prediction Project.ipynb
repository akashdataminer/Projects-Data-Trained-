# Splitting the Dataset into Train & Test:-

```python
x=doctor_new.drop(columns=["Fees"])
x

# Splitting all the Independent variable in x
```

```python
y=doctor_new.Fees
y

# Splitting the target variable in y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
s=StandardScaler()
x=s.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import r2_score

# Importing Linear Regression and Evolution Matrix
```

```python
MaxAccu=0
MaxRS=0

for i in range (1,500):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=i)
    abr=AdaBoostRegressor()
    abr.fit(x_train,y_train)
    predabr_train=abr.predict(x_train)
    predabr_test=abr.predict(x_test)
    acc=r2_score(y_test,predabr_test)
    if acc>MaxAccu:
        MaxAccu=acc
        MaxRS=i
print("Best Accuracy is ", MaxAccu, "on Random State ", MaxRS)


# The Best Random State with Best Accuracy
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=145)
```

```python
abr=AdaBoostRegressor()
abr.fit(x_train,y_train)

# Training the Adaboost Regressor Model
```

```python
abr_pred=abr.predict(x_test)
abr_pred

# The Adaboost Regressor Model is predicting values based on training
```

```python
print(r2_score(y_test,abr_pred))

# The R2 Accuracy Score of Adaboost Regressor Model
```

# Cross Validation:-

```python
from sklearn.model_selection import cross_val_score

# Importing Cross val Score to perform Cross Validation
```

```python
scr=cross_val_score(abr, x, y, cv=5)
scr
```

```python
print("The Cross Validation Score of Adaboost Regressor is :", scr.mean())

# The Cross Validation Score of Adaboost Regressor is : 0.11863785662220946
```
