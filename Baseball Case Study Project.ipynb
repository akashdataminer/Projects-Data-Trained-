# Data Cleaning:-

### Outliers Treatment:-

```python
from scipy.stats import zscore

# Importing the Zscore from the Scipy Library
```

```python
z=np.abs(zscore(baseball))
z

# We are taking the absolute of the zscore so all the values will be displayed in positive standard Deviation
```

```python
print(np.where(z>3))

# Printing the integer location of all the values that have Standard Deviation greater than +3
# These points determine the point of outliers
```

```python
baseball_new=baseball[(z<3).all(axis=1)]
baseball_new

# Taking all the values within +3 Standard Deviation and saving it in another DataFrame "baseball_new"
# This new Dataset formed is without the outliers
```

```python
baseball.shape
```

```python
baseball_new.shape
```

```python
data_loss=((30-29)/30)*100
data_loss

# The Data Loss is 3.33% which is less than 5%, Therefore we will drop the outliers
```

# Splitting the Model for Train & Test:-

```python
x=baseball_new.iloc[:,1:]
x

# Taking all the independent variables into x
```

```python
y=baseball_new.iloc[:,0]
y

# Taking the target variable into y
```

### Building Model using Linear Regression:-

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# Importing the Algorithms and Evolution Matrix
```

### Finding Best Random State:-

```python
MaxAccu=0
MaxRS=0

for i in range (1,500):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=i)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    predlr_train=lr.predict(x_train)
    predlr_test=lr.predict(x_test)
    acc=r2_score(y_test,predlr_test)
    if acc>MaxAccu:
        MaxAccu=acc
        MaxRS=i
print("Best Accuracy is ", MaxAccu, "on Random State ", MaxRS)

# Finding the Best Random State with Maximum Accuracy
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=6)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)

# Training the Linear Regression Model
```

```python
lr_pred=lr.predict(x_test)
lr_pred

# The Linear Regression Model predicting the values based on training
```

```python
print(r2_score(y_test,lr_pred))

# The R2 Accuracy Score of Linear Regression Model
```

```python
print(mean_absolute_error(y_test,lr_pred))
print(mean_squared_error(y_test,lr_pred))
print(np.sqrt(mean_squared_error(y_test,lr_pred)))

# Evolution MAtrix of Linear Regression Model
```

```python
GCV.best_params_

# Getting the Best Parameters with Best Accuracy at {'criterion': 'mse',
#  'max_depth': 2,
#  'min_samples_leaf': 4,
#  'min_samples_split': 9,
#  'n_estimators': 3}
```
