### Verifying Class Imbalance Problem:-

```python
x=hr_analytics.drop(columns=["Attrition"])
x

# Taking all independent variables in x
```

```python
y=hr_analytics.Attrition
y

# Taking dependent variable in y
```

```python
hr_analytics["Attrition"].hist(grid=True)
plt.title("HR Attrition Graph")
plt.show()

# There is a class imbalance problem as there are many people who are not part of turnover and very few people leave company
```

```python
from imblearn.over_sampling import SMOTE

# Importing SMOTE function to treat the class imbalance problem with Over Sampling method
```

```python
sm=SMOTE()
```

```python
x_over, y_over = sm.fit_resample(x,y)

# Resampling the x and y to make the class balanced and saving the new x and y in x_over and y_over respectively
```

```python
y_over.value_counts()
```

```python
y_over.hist(grid=True)
plt.title("HR Attrition Graph Revised")
plt.show()

# We can see that now class imbalance problem is treated and we have equal number of samples present
```

### Finding Best Random State:-

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split

# Importing the algorithms and Evolution Matrix
```

```python
MaxAccu=0
MaxRS=0

for i in range (1,500):
    x_train,x_test,y_train,y_test=train_test_split(x_over,y_over,test_size=0.20,random_state=i)
    lr=LogisticRegression()
    lr.fit(x_train,y_train)
    predlr_train=lr.predict(x_train)
    predlr_test=lr.predict(x_test)
    acc=accuracy_score(y_test,predlr_test)
    if acc>MaxAccu:
        MaxAccu=acc
        MaxRS=i
print("Best Accuracy is ", MaxAccu, "on Random State ", MaxRS)
```

### Model Building using Logistic Regression:-

```python
x_train,x_test,y_train,y_test=train_test_split(x_over,y_over,test_size=0.20,random_state=325)
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)

# Training Logistic Regression Model
```

```python
lr_pred=lr.predict(x_test)
lr_pred

# The Logistic Regression Model is predicting values based on training
```

```python
print(accuracy_score(y_test,lr_pred))

# The Accuracy Score of Logistic Regression Model
```

```python
print(confusion_matrix(y_test,lr_pred))
print(classification_report(y_test,lr_pred))

# Evolution Matrix of Logistic Regression Model
```

# Hyper Parameter Tuning:-

```python
from sklearn.model_selection import GridSearchCV

# Importing the GridSearchCV to get the best parameters of the Extra Trees Classifier model
```

```python
parameter={"max_depth":np.arange(2,10), 
           "criterion":["gini","entropy"],
          "min_samples_split":np.arange(2,10),
          "n_estimators":np.arange(2,10),
          "min_samples_leaf":np.arange(2,10)}

# Setting the Parameters to apply to GridSearchCV to get the best parameter score
```

```python
GCV.best_params_

# The best parameter we got is at :- 
# ['criterion': 'entropy', 'max_depth': 9, 'min_samples_leaf':2,'min_samples_split': 2, 'n_estimators': 8 ]
```
